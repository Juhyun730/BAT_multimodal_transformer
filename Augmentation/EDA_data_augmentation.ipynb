## DATA augmentation

```python
list_col = data['EDA']
list_col_array = np.array(list_col.tolist())
EDA_new = pd.DataFrame(list_col_array, columns=[f'col{i+1}' for i in range(list_col_array.shape[1])])
```

```python
result = result.apply(pd.to_numeric, errors='coerce')
result = result.dropna()
```

## Train,Valid data set 나누기

```python
n_samples = result.shape[0]
n_group1 = int(n_samples * 0.875)
n_group2 = n_samples - n_group1

group1 = result.sample(n=n_group1)
group2 = result.drop(group1.index).sample(n=n_group2)
```

```python
print(group1.shape)  
print(group2.shape)  
```

```python
selected_zero = group1[group1['sentiment_x'].isin([0])]
```

```python
# rain data set 안에 감정 상태 : 중립
selected_zero
```

```python
# Train data set 안에 중립을 제외한 감정
selected_rows = group1[group1['sentiment_x'].isin([1,2,3,4,5,6])]
```

```python
n_samples = 3500
selected_zero_sample = selected_zero.sample(n=n_samples)
```

## 중립이 아닌 감정 증강

```python
selected_rows = selected_rows.astype('float64')
```

```python
X = selected_rows.drop(['sentiment_x'], axis=1).values
y = selected_rows['sentiment_x'].values
```

```python
smote = SMOTE(sampling_strategy={1:3500,2: 3500, 3: 3500, 4: 3500, 5: 3500, 6: 3500},random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)
```

```python
X_resampled
```

```python
y_resampled_df = pd.DataFrame(y_resampled, columns=['sentiment_x'])
```

```python
df = pd.DataFrame(np.zeros((1, 80)), columns=['col'+str(i) for i in range(1, 81)])
```

```python
X_resampled_df = pd.DataFrame(X_resampled, columns=df.columns)
```

```python
X_resampled_df.head()
```

```python
data_augmented = pd.concat([X_resampled_df, y_resampled_df], axis=1)
```

```python
data_augmented = pd.concat([data_augmented, selected_zero_sample ], axis=0)
```
